```python
%%sql
create table Vehiclez (
LicencePlate VARCHAR(15) primary key,
Make text,
Model text,
year INT,
Color text,
VIN VARCHAR(17)
);
```

```python
%%sql
create table Driver (
FirstName text,
LastName text,
LicenseNumber VARCHAR(20),
Phone VARCHAR(15),
Address VARCHAR(100),
City text,
State text,
ZipCode VARCHAR(10)
 );
```

```python
%%sql
create table Trip (
VehicleID INT,
DriverID INT,
StartDate DATE,
EndDate DATE,
StartLocation text,
EndLocation text,
Distance VARCHAR
);
```

```python
%%sql
create table Maintenances (
VehicleID INT,
MaintenanceDate DATE,
Description text,
cost DECIMAL
);
```
