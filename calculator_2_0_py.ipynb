```python
import math

# Create a class called "Calculator"
class Calculator:
    def __init__(self):  # Initializes the dictionary with basic mathematical operations
        # Dictionary attribute to store the available mathematical operations and their corresponding functions
        self.operations = {
            '+': lambda x, y: x + y,
            '-': lambda x, y: x - y,
            '*': lambda x, y: x * y,
            '/': lambda x, y: x / y if y != 0 else float('inf')  # Handle division by zero
        }

    # Method called "add_operation" that adds the new operation and function to the dictionary.
    def add_operation(self, symbol, function):
        self.operations[symbol] = function

    # Method called "calculate" that uses the dictionary to determine the appropriate function to perform the calculation
    def calculate(self, num1, operation, num2):
        # Check if inputs are numbers
        if not (isinstance(num1, (int, float)) and isinstance(num2, (int, float))):
            raise ValueError("Inputs must be numbers.")

        # Check if the operation symbol is valid
        if operation not in self.operations:
            raise ValueError(f"Operation '{operation}' is not supported.")

        # Perform calculations
        return self.operations[operation](num1, num2)

# Advanced mathematical operations
def exponentiation(x, y):
    return x ** y

def square_root(x, _=None):
    if x < 0:
        raise ValueError("Cannot calculate the square root of a negative number.")
    return math.sqrt(x)

def logarithm(x, _=None):
    if x <= 0:
        raise ValueError("Logarithm undefined for non-positive values.")
    return math.log(x)

if __name__ == "__main__":
    calc = Calculator()

    # Add advanced operations
    calc.add_operation('^', exponentiation)
    calc.add_operation('sqrt', square_root)
    calc.add_operation('log', logarithm)

    # Create an instance of the Calculator class
    # Use a while loop that allows the user to continue performing calculations until they choose to exit
    while True:
        try:
            # Get user input
            num1 = input("Enter the first number (or type 'exit' to quit): ")
            if num1.lower() == 'exit':
                print("Goodbye!")
                break

            # Convert num1 to a float
            num1 = float(num1)

            # Get operation
            operation = input("Enter the operation (+, -, *, /, ^, sqrt, log): ")

            # Get second number if the operation requires it
            num2 = 0
            if operation not in ['sqrt', 'log']:
                num2 = float(input("Enter the second number: "))

            # Perform calculation
            result = calc.calculate(num1, operation, num2)
            print(f"Result: {result}\n")

        except ValueError as e:
            print(f"Error: {e}\n")
        except Exception as e:
            print(f"An unexpected error occurred: {e}\n")
```

```python
import math

class Calculator:

  def __init__(self):

    self.operations = {
      "+": self.add,
      "-": self.subtract,
      "*": self.multiply,
      "/": self.divide,
    }

  def add_operation(self, symbol, function):

    self.operations[symbol] = function

  def calculate(self, num1, symbol, num2):

    if symbol not in self.operations:
      raise ValueError(f"Invalid operation: {symbol}")

    if not isinstance(num1, (int, float)) or not isinstance(num2, (int, float)):
      raise ValueError("Inputs must be numbers")

    return self.operations[symbol](num1, num2)

  def add(self, num1, num2):

    return num1 + num2

  def subtract(self, num1, num2):

    return num1 - num2

  def multiply(self, num1, num2):

    return num1 * num2

  def divide(self, num1, num2):

    if num2 == 0:
      raise ZeroDivisionError("Division by zero")
    return num1 / num2

def exponent(num1, num2):

  return math.pow(num1, num2)

def square_root(num):

  if num < 0:
    raise ValueError("Cannot take square root of a negative number")
  return math.sqrt(num)

def logarithm(num, base=10):

  if num <= 0:
    raise ValueError("Logarithm of non-positive number is undefined")
  return math.log(num, base)

# Add advanced operations to the calculator
calculator = Calculator()
calculator.add_operation("^", exponent)
calculator.add_operation("sqrt", square_root)
calculator.add_operation("log", logarithm)

# Main program loop
while True:
  try:
    num1 = float(input("Enter the first number: "))
    symbol = input("Enter the operation (+, -, *, /, ^, sqrt, log): ")
    num2 = float(input("Enter the second number: ")) if symbol not in ["sqrt", "log"] else float(input("Enter the number (for sqrt and log): "))

    result = calculator.calculate(num1, symbol, num2)
    print(f"{num1} {symbol} {num2} = {result}")

  except (ValueError, ZeroDivisionError) as e:
    print(f"Error: {e}")

  # Ask if user wants to continue
  choice = input("Do you want to continue (y/n)? ")
  if choice.lower() != "y":
    break

print("Thank you for using the calculator!")
```
