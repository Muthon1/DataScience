```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sn
```

```python
climate_change_Africa='/content/drive/MyDrive/Checkpoint Assignments/Africa_climate_change.csv'
df=pd.read_csv(climate_change_Africa)
df.head()
```

```python
# checking for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# dropping rows with missing values
df_cleaned = df.dropna()
print(df_cleaned)
```

```python
# Forward fill (propagate previous value)
df_filled_ffill = df.fillna(method='ffill')
print(df_filled_ffill)
```

```python
# Backward fill (propagate next value backward)
df_filled_bfill = df.fillna(method='bfill')
print(df_filled_bfill)
```

```python
# Dropping rows with missing values in 'TAVG' column only
df_cleaned = df.dropna(subset=['TAVG'])
print(df_cleaned)
```

```python
# Remove outliers beyond a certain temperature threshold (e.g., min temp below -10 or max temp above 50)
df_cleaner = df[(df['TMIN'] >= -10) & (df['TMAX'] <= 100)]
print(df_cleaner)
```

```python
# Plot a line chart to show the average temperature fluctuations in Tunisia and Cameroon
import matplotlib.pyplot as plt

# Filter the data for Tunisia and Cameroon
df_filtered = df[(df['COUNTRY'].isin(['Tunisia', 'Cameroon']))]

# Plot the line chart for average temperature fluctuations
plt.figure(figsize=(10,6))

# Loop through both countries to plot their data
for country in ['Tunisia', 'Cameroon']:
    country_data = df_filtered[df_filtered['COUNTRY'] == country]
    plt.plot(country_data['TAVG'], label=country)

# Step 3: Customize the chart
plt.title('Average Temperature Fluctuations: Tunisia vs Cameroon', fontsize=14)
plt.xlabel('DATE', fontsize=12)
plt.ylabel('Average Temperature (Â°C)', fontsize=12)
plt.legend(title="COUNTRY")
plt.grid(True)

# Step 4: Display the plot
plt.show()
```

```python
# Zoom in to only include data between 1980 and 2005
df_filtered = df[(df['DATE'] >= '1980-01-01') & (df['DATE'] <= '2005-12-31')]

# Plot the line chart for average temperature fluctuations
plt.figure(figsize=(10,6))

# Loop through both countries to plot their data
for country in ['Tunisia', 'Cameroon']:
    country_data = df_filtered[df_filtered['COUNTRY'] == country]
    plt.plot(country_data['TAVG'], label=country)
plt.show()
```

```python
# Create Histograms to show temperature distribution in Senegal between [1980,2000] and [2000,2023]
df_filtered = df[(df['DATE'] >= '1980-01-01') & (df['DATE'] <= '2000-12-31')]
plt.figure(figsize=(10,6))
plt.hist(df_filtered['TAVG'], bins=20, color='blue', alpha=0.7)
plt.title('Temperature Distribution in Senegal (1980-2000)and (2000-2023)')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
