#Understanding the purpose of each NLP(Natural Language Processing) technique
i) Tokenization:This is the process of breaking text into smaller units known as tokens. they are basically words or subwords.

ii) Part-of-speech Tagging: Refers to the identification of the grammatical category of each word in a sentence. For instance, is it a verb, adjective etc.

iii) Named Entity Recognition: Process of extracting named entities such as people, organizations and places from text.

iv) Sentiment analysis: This is basically determining the sentiment or emotion behind a text. It can either be positive, negative or neutral

v) Text Classification:This is categorization of texts into predefined classes or labels. It's commonly used in spam detection, topic classification or language detection.

vi) Dependency parsing: It's the process of analyzing the grammatical structure of a text to identify the relationships between words.

vii) Stemming and Lemmatization: serves the purpose of reducing words to their base form such as jumping to 'jump'. Stemming employs algorithms that remove prefixes or suffixes.

viii) Word Embeddings: Representing words in numerical form so that words with similar meanings are close in the vector space.

ix) Machine Translation: A piece of text is translate from one language to another automatically.

x) Text Summarization: Automatically condensing a long document or text into a shorter version.

xi)Topic Modeling: Discovering hidden topics that are in a collection of texts.

xii) Speech Recognition: Changing spoken language into written text.

xiii) Text Generation: Generating coherent text from a given input.

xiv) Coreference Resolution: Determining which words in a sentence refer to the same entity.

xv) Question Answering: Generating a response to a question from a given knowledge base.

```python
# Tokenizing the text

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download necessary NLTK resources
nltk. download('stopwords')
nltk. download('punkt')
nltk. download('punkt_tab')

# Tokenize the first entry
processed_texts = [word_tokenize(text) for text in texts]
tokens = word_tokenize(texts.iloc[0])
```

```python
# Part-of-Speech Tagging
from nltk import pos_tag

nltk.download('averaged_perceptron_tagger_eng')

# POS tagging for the first processed text
tagged_tokens = pos_tag(processed_texts[0])
print(tagged_tokens)
```

```python
# Named Entity Recognition
from nltk import ne_chunk

nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

# Named Entity Recognition for the first processed text
entities = ne_chunk(pos_tag(processed_texts[0]))
print(entities)
```

```python
# Sentiment Analysis
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

# Initialize SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
sentiment_scores = [sia.polarity_scores(text) for text in texts]
print(sentiment_scores)
```

```python
# Text Classification
import sklearn
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

# Download necessary NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

# Clean the text
def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)  # Remove URLs
    text = re.sub(r'@\w+', '', text)  # Remove mentions (e.g., @user)
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove non-alphabetic characters
    return text

cleaned_texts = [clean_text(text) for text in texts]

# Convert cleaned texts into TF-IDF features
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
X = tfidf_vectorizer.fit_transform(cleaned_texts)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, df['tweet_id'], test_size=0.2, random_state=42)

#Train a Naive Bayes Classifier
classifier = MultinomialNB()
classifier.fit(X_train, y_train)

# Make predictions
y_pred = classifier.predict(X_test)

# Evaluate the model
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Dependency Parsing
# Necessary libraries
from nltk import CFG
from nltk.parse import ChartParser

# sample processed text
processed_texts =[["the", "opened", "causing", "reply", "under"]]
processed_texts_flat = [word for sublist in processed_texts for word in sublist]

# Removing stopwords
stop_words = ['to', 'be', 'and', 'is']
processed_texts_filtered = [word for word in processed_texts_flat if word.lower() not in stop_words]


# Handling the emojis
processed_texts_final = [word if not word.startswith('@') and not '😡' in word else 'opened' for word in processed_texts_filtered]

# Define simple grammar
grammar = CFG.fromstring("""
  S -> NP VP
  VP -> V NP | V NP PP | V | V ADJP
  NP -> Pronoun | ProperNoun | Det N | Det N PP | ProperNoun 'to' V
  ADJP -> Adj
  Pronoun -> 'She' | 'He' | 'It' | 'They'
  ProperNoun -> 'James' | '@AppleSupport'
  Det -> 'the' | 'a' | 'an'
  N -> 'reply' | 'notification' | 'keyboard' | 'app' | 'phone' | 'issue'
  V -> 'causing' | 'see' | 'liked' | 'opened' | 'disregarded' | 'tapped' | 'is' | 'be'
  PP -> 'under' NP | 'to' NP | 'by' NP | 'with' NP
  Adj -> 'opened'
""")

# Initialize the parser
parser = ChartParser(grammar)

# Parse the first processed text
try:
    parsed_tree = parser.parse(processed_texts_final)
    for tree in parsed_tree:
        print(tree)
except ValueError as e:
    print(f"Error: {e}")
```

```python
# Stemming and Lemmatization
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet

# Download necessary NLTK resources
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize Stemmer and Lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Example words
words = ["causing", "tapped", "disregarded", "be"]

# Stemming
stemmed_words = [stemmer.stem(word) for word in words]
print("Stemmed Words:", stemmed_words)

# Lemmatization
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in words]
print("Lemmatized Words:", lemmatized_words)
```

```python
# Text Summarization
from nltk import sent_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.tokenize.treebank import TreebankWordDetokenizer

# Download necessary NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

# Sample text
text = "Your business means a lot to us. Please DM your name, zip code and additional details about your concern"

# Tokenization and preprocessing
stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(text)
filtered_words = [word.lower() for word in word_tokens if word.isalpha() and word.lower() not in stop_words]

# Frequency Distribution
fdist = FreqDist(filtered_words)

# Detokenize summary
summary = TreebankWordDetokenizer().detokenize([word for word in filtered_words if fdist[word] > 1])
print("Summary:", summary)
```

```python
#Topic Modelling
from gensim import corpora
from gensim.models import LdaModel

# Example documents
documents = ["Your business means a lot to us", "LiveChat is online at the moment", "Battery life is important, and we're here for you"]

# Tokenize and prepare corpus
texts = [doc.split() for doc in documents]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# Train LDA model
lda = LdaModel(corpus, num_topics=2, id2word=dictionary, passes=15)

# Print topics
for topic in lda.print_topics():
    print(topic)
```

```python
# Text Generation
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Encode input and generate text
input_text = "Your business means a lot to us"
inputs = tokenizer.encode(input_text, return_tensors='pt')
outputs = model.generate(inputs, max_length=50, num_return_sequences=1)

# Decode generated text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Generated Text:", generated_text)
```

```python
# Coreference Resolution
import spacy

# Load the spacy model with coreference resolution
nlp = spacy.load("en_core_web_sm")
doc = nlp("Battery life is important, and we're here for you. DM us the iOS version your iPhone is running.")

#coreference resolution
for np in doc:
    if np.dep_ == 'nsubj' and np.head.pos_ == 'VERB':
        print(f'{np.text} refers to {np.head.text}')
```
