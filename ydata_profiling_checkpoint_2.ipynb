```python
# correlations between different columns in the dataset
import pandas as pd

# Assuming 'Flight_date' is the column with dates
# Convert 'Flight_date' to datetime objects
Tunisair_dataset['Filght_date'] = pd.to_datetime(Tunisair_dataset['Filght_date'])

# Select only numerical columns for correlation calculation
numerical_columns = Tunisair_dataset.select_dtypes(include=['number']).columns

# Calculate correlations using only numerical columns
correlations = Tunisair_dataset[numerical_columns].drop(columns=['Filght_date'], errors='ignore').corr()

print(correlations)
```
